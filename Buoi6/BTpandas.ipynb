```python
import pandas as pd
import numpy as np
df = pd.read_csv('categorical_data.csv')
df.head()
```

1.Kiểm tra các missing data

```python
df.isnull().sum()
```

Cách 1: Thay thế các ô missing data bằng các giá trị trung bình mean()

```python
df['age'].fillna(df['age'].mean(), inplace = True)
```

```python
df['age']
```

```python
df['salary'].fillna(df['salary'].mean(), inplace = True)
```

```python
df['salary']
```

Cách 2 : sử dụng SimpleImpute

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values=np.NaN, strategy='mean')
df.age = imputer.fit_transform(df['age'].values.reshape(-1,1))[:]
```

```python
df['age']
```

```python
df.salary = imputer.fit_transform(df['salary'].values.reshape(-1,1)[:])
```

```python
df['salary']
```
