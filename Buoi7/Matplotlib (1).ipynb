```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Mall_Customers_2.csv")
#đọc file
```

```python
df.head(200)
```

```python
df.isnull().sum()
#kiểm tra missing data
```

```python
df['Age'].fillna(df['Age'].mean(),inplace = True)
df['Spending Score (1-100)'].fillna(df['Spending Score (1-100)'].mean(),inplace = True)
#xử lí missing data bằng giá trị mean() : trung bình
```

```python
df = df.sort_values(by='Age')
```

```python
tuoi,diem = df['Age'] , df['Spending Score (1-100)']
#gán tuoi bằng cột Age và diem bằng cột Spending....
```

```python
color = ['blue' if i == 'Male' else 'pink' for i in df['Genre']]
#gán màu cho Male là blue và ko phải Male là Female hay bê đê là hồng
```

```python
for i in range(len(color)) :
    plt.scatter(tuoi[i],diem[i],color = color[i])
plt.xlabel('Tuổi')
plt.ylabel("Điểm tiêu dùng")
#vẽ biều đồ, gán nhãn cho cái trục hoành và trục tung
```

```python
luong = df['Annual Income (k$)']
#gán luong bằng cột Annual...
plt.plot(tuoi,luong)
plt.annotate('Max',xy=(31,137),xytext=(41,147),arrowprops = dict(facecolor = 'black'))
#set cái điểm max
```

```python
p2 = pd.read_csv("Social_Network_Ads_2.csv");
p2.head()
```

```python
p2.isnull().sum()
#ktra missing data
```

```python
p2['Purchased'].fillna(p2['Purchased'].mean(),inplace = True)
p2['EstimatedSalary'].fillna(p2['EstimatedSalary'].mean(),inplace = True)
#xử lí missing data
p2.head()
```

```python
p2['Purchased'] = p2['Purchased'].apply(lambda i : 'Chưa mua' if i == 1 else 'Đã mua')
```

```python
p2['Status'] = p2['Purchased']
p2 = p2.drop('Purchased',axis =1)
p2.head()
```

```python
tuoi = p2['Age']
luong = p2['EstimatedSalary']
color = ['red' if i == 'Chưa mua' else 'blue' for i in p2['Status']]
for i in range(len(color)) :
  plt.bar(tuoi[i],luong[i],color = color[i])
```

```python
color = ['Purple' if i == 'Đã mua' else 'red' for i in p2['Status']]
for i in range(len(color)) : 
  plt.scatter(tuoi[i],luong[i],color = color[i])
```
